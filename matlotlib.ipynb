```python
import pandas as pd
df_sales=pd.read_excel("linechart.xlsx")
df_sales.head()
```

```python
from matplotlib import pyplot as plt
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"])
```

```python

plt.figure(figsize=(12,5))
plt.plot(df_sales["Quarter"],df_sales["Fridge"])
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],label="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Fridge"],label="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],label="Washing Machine")

plt.title("PRODUCT SALES")
plt.ylabel("Revenue(INR)")
plt.xlabel("Quarterly")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,'+--',color='red',label="max")
plt.plot(days,min_t,'--d',label="min")
plt.plot(days,avg_t,'--*',label="avg")
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title('weather')
plt.legend(loc='best',shadow=True,fontsize="small")
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xpos=np.arange(len(company))
```

```python

plt.title("mncs")
plt.xlabel("company")
plt.ylabel("revenue")
plt.xticks(xpos,company)
plt.legend()

plt.bar(xpos -0.2,revenue,width=0.3,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.3,label="Profit")
plt.legend()
```
